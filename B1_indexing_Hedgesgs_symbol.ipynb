### Additional ENSMBL GIDs and descriptions are dropped on merging data, so import and add this data again

```python
dfX=pd.read_csv('../BioMart_Ensmbl_index/mart_export72_allMGIsymbol.txt',index_col=[3]) 
dfX.head() 
```

```python
dfX.shape
```

```python
dfX[dfX.index.duplicated()==True]  #showing duplicate entries exist in the key
```

```python
dfB.shape
```

```python
df_FINAL = dfB.join(dfX, how='left', sort=True)
df_FINAL[df_FINAL.index.duplicated()==True]  #showing duplicate entries still exist (or use .shape for number)
```

```python
dfOUT =df_FINAL.groupby(df_FINAL.index).mean()

dfOUT.shape   #check number of individual GIDs
```
